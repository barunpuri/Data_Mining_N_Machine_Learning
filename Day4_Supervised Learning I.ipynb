# Supervised Learning I

Supervised Learning은 input과 정답이 있는 데이터 집합을 학습시키는 방법이다. input과 output의 관계를 유추하여 올바른 해답을 제시한다. 사람이 직접 labeling을 하여 정확도가 높은 데이터를 사용할 수 있다.

지도학습에는 Regression과 Classification이 있다. 
Regression은 input과 output을 분석하여 잘 표현하는 연속 함수를 찾는 과정이며, Classification은 input에 따른 이산적인 output값을 찾는다. 

실제 raw data를 가공하여 특징을 찾아낸 후, 데이터의 특징을 잘 표현하는 Model생성, 생성된 Model로 부터 결과를 예측하고 이를 분석하는 과정을 Feature Engineering이라고 한다.

```python
import pandas as pd
```

```python
df = pd.read_csv('data/retail_transaction_data.csv')
df.head(10)
```

```python
df.shape
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['revenue'] = df['UnitPrice']*df['Quantity']
```

```python
operations = {'revenue':'sum','InvoiceDate':'first', 'CustomerID':'first'}
df = df.groupby('InvoiceNo').agg(operations)
```

```python
df.head(10)
```

```python
df['year'] = df['InvoiceDate'].apply(lambda x: x.year)
df.head(10)
```

```python
df['days_since'] = (pd.datetime(year=2010, month=12, day=31) - df['InvoiceDate']).apply(lambda x: x.days)
df.head(10)
```

```python
operations = {'revenue':'sum',
              'days_since':['max','min', 'nunique'],
             }
X = df[df['year'] == 2010].groupby('CustomerID').agg(operations)
X.head()
```

```python
X.columns = [' '.join(col).strip() for col in X.columns.values]
X.head()
```

```python
X['average_order_cost'] = X['revenue sum']/X['days_since nunique']
X.head()
```

```python
Y = df[df['year'] == 2011].groupby('CustomerID')['revenue'].sum()
Y.head()
```

```python
wrangled_df = pd.concat([X,Y], axis=1)
wrangled_df.columns = ['2010 revenue',
                       'days_since_first_purchase',
                       'days_since_last_purchase',
                       'number_of_purchases',
                       'avg_order_cost',
                       '2011 revenue']
wrangled_df.head(10)
```

```python
wrangled_df = wrangled_df[~wrangled_df['2010 revenue'].isnull()]
wrangled_df = wrangled_df[~wrangled_df['2011 revenue'].isnull()]
wrangled_df.head()
```

```python
print(wrangled_df['2011 revenue'].median() + wrangled_df['2011 revenue'].std()*3)
print(wrangled_df['2010 revenue'].median() + wrangled_df['2010 revenue'].std()*3)
```

```python
wrangled_df = wrangled_df[wrangled_df['2011 revenue'] < ((wrangled_df['2011 revenue'].median()) + wrangled_df['2011 revenue'].std()*3)]
wrangled_df = wrangled_df[wrangled_df['2010 revenue'] < ((wrangled_df['2010 revenue'].median()) + wrangled_df['2010 revenue'].std()*3)]
wrangled_df.head()
```

```python
wrangled_df.to_csv('Data/wrangled_transaction_data_result.csv')
wrangled_df.head(100)
```
